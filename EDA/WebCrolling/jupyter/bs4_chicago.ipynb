```python
from urllib.parse import urljoin
import re

rank_list = []
menu_list = []
cafe_list = []
link_list = []

for item in soup.find_all('div', 'sammy'):
    rank_list.append(item.find(class_= "sammyRank").string)
    text_list = re.split(('\n'),item.find(class_ = "sammyListing").get_text())
    menu_list.append(text_list[0])
    cafe_list.append(text_list[1])

    link_list.append(urljoin("https://www.chicagomag.com", item.find('a')['href']))
```

```python
for i in range(5):
    print(rank_list[i] + '. ' + menu_list[i] + " - " + cafe_list[i] + " " + link_list[i])
```

```python
import pandas as pd

data = {'Rank':rank_list, 'Menu' : menu_list, 'Cafe' : cafe_list, 'URL':link_list}
df = pd.DataFrame(data)
```

```python
from tqdm import tqdm

import pandas as pd

data = {'Rank':rank_list, 'Menu' : menu_list, 'Cafe' : cafe_list, 'URL':link_list}
df = pd.DataFrame(data)

print(df)
req = Request(url = df['URL'][0], headers={'User-Agent': generate_user_agent()})
page = urlopen(req)
soup = BeautifulSoup(page, "html.parser")

#print(soup.prettify())
price_list = []
address_list = []
cafenum_list = []

for idx, row in tqdm(df.iterrows(), total=len(df.index)):

    req = Request(url = row['URL'], headers = {'User-Agent':generate_user_agent()})
    page = urlopen(req)
    soup = BeautifulSoup(page, "html.parser")

    text = re.split(".,", soup.find('p', 'addy').get_text())[0]
    price = re.search('\$\d+\.(\d+\.)?', text).group()
    price=price.strip("$")
    price=price[:-1]
    

    address = text[len(price)+3:]
    print(address)
    
    cafenum = re.split(".,", soup.find('p', 'addy').get_text())[1]
    #for i in range(len(cafenum)):
    #    print(cafenum[i])
    #    print(type(cafenum[i]))
    #homepage = re.split(".,", soup.find('p', 'addy').get_text())[2]
    
    price_list.append(float(price))
    
    if re.search("\d+\-\d+\-\d+", cafenum):
        cafenum_list.append(cafenum)
    elif re.search("\w+\.\w+", cafenum):
        cafenum_list.append('X')
    else:
        #print(address)
        #print(type(address))
        #print(cafenum)
        address=address + cafenum
        #print(address)
        cafenum_list.append('X')

    address_list.append(address)
#price_list, address_list
cafenum_list
```

```python
len(price_list), len(address_list)
```

```python
df['Price'] = price_list
df['Address'] = address_list
df['cafenum'] = cafenum_list
#df = df.loc[:, ['Rank', 'Cafe', 'Menu', 'Price', 'Address', 'cafenum']]
df.set_index('Rank', inplace = True)
df.head()
```
