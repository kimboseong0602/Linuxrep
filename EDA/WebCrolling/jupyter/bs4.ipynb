```python
import urllib.parse
from bs4 import BeautifulSoup
from urllib.request import urlopen

import urllib
from urllib.request import Request

url = "https://search.naver.com/search.naver?query={search_word1}+{search_word2}"

req = Request(url.format(search_word1 = urllib.parse.quote("박스오피스"),
                         search_word2 = urllib.parse.quote("평점순")))

res = urlopen(req)
soup = BeautifulSoup(res, "html.parser")
print(soup.prettify())
```

```python
movie_panel = soup.find_all(class_ = "_panel")
movie_panel
```

```python
rating = soup.find(class_="_panel_rating _tab_content")
for each in rating.find_all("li"):
    print(each.prettify())
```

```python
rating = soup.find(class_="_panel_rating _tab_content")
for idx, each in enumerate(rating.find_all("li")):
    print("[" + str(idx) + "]")
    print("title : " + each.find("strong", class_ = "name").string)
```

```python
rating = soup.find(class_="_panel_rating _tab_content")
for idx, each in enumerate(rating.find_all("li")):
    print("[" + str(idx) + "]")
    print("title : " + each.find("strong", class_ = "name").string)
    print("rating : " + each.find("span", class_ = "sub_text").get_text())
```

```python
rating = soup.find(class_="_panel_rating _tab_content")
for idx, each in enumerate(rating.find_all("li")):
    print("[" + str(idx) + "]")
    print("title : " + each.find("strong", class_ = "name").string)
    print("rating : " + each.find("span", class_ = "sub_text").get_text())
    print("link : " + each.find("a")["href"])
```

```python
prefix = "https://search.naver.com/search.naver"
movie_list = []

rating = soup.find(class_ = "_panel_rating _tab_content")
for each in rating.find_all("li"):
    movie = []
    movie.append(each.find("strong", class_ = "name").string)
    movie.append(each.find("span", class_ = "sub_text").get_text())
    movie.append(prefix + each.find("a")["href"])

    movie_list.append(movie)

movie_list
```

```python
res = urlopen(movie_list[13][2])
soup = BeautifulSoup(res, "html.parser")
print(soup.find("strong", class_ = "_text").string)
```

```python
def get_title(url):
    res = urlopen(url)
    soup = BeautifulSoup(res , "html.parser")
    return soup.find("strong", class_ = "_text").string
```

```python
get_title(movie_list[13][2])
```

```python
for movie in movie_list:
    movie[0] = get_title(movie[2])

movie_list
```

```python
import time

for movie in movie_list:
    movie[0] = get_title(movie[2])
    time.sleep(0.5)

movie_list
```
