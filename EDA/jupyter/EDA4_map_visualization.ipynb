```python
import json
import pandas as pd

state_data = pd.read_csv("~/class/dev_ws/EDA/data/02. US_Unemployment_Oct2012.csv")

m = folium.Map(location = [48, -102], zoom_start=3)

folium.Choropleth(
    geo_data='../data/02. us-states.json',
    data = state_data,
    columns = ['State', 'Unemployment'],
    key_on = 'feature.id',
    fill_color= 'YlGn',
    fill_opacity=0.7,
    line_opacity = 0.2,
    legend_name='Unemployment Rate(%)'
).add_to(m)
```

```python
crime_anal_norm = pd.read_csv('../data/seoul_crime_result.csv', index_col = 0, encoding = 'utf-8')

geo_path = '../data/02. skorea_municipalities_geo_simple.json'

geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
my_map = folium.Map(location=[37.5502, 126.982], zoom_start=11)

folium.Choropleth(
    geo_data=geo_str,
    data = crime_anal_norm['살인'],
    columns=[crime_anal_norm.index, crime_anal_norm['살인']],  #자치구별 인덱스를 컬럼으로
    fill_color='PuRd',
    key_on = 'feature.id',
    fill_opacity = 0.7,
    line_opacity = 0.2,
    legend_name = '정규화된 살인 발생 건수'
).add_to(my_map)
```

```python
my_map
```

```python
my_map = folium.Map(location=[37.5502, 126.982], zoom_start=11)

folium.Choropleth(
    geo_data = geo_str,
    data = crime_anal_norm['성범죄'],
    columns=[crime_anal_norm.index, crime_anal_norm['성범죄']],
    fill_color='PuRd',
    key_on='feature.id',
    fill_opacity=0.7,
    line_opacity=0.2,
    #opacity는 투명도
    legend_name="정규화된 성범죄 발생 건수"
).add_to(my_map)
```
