```python
import numpy as np
crime_station=raw_data.pivot_table(index='구분', columns=['죄종', '발생검거'], aggfunc=np.sum, fill_value=0)

crime_station.head()
```

```python
crime_station.columns = crime_station.columns.droplevel([0])
```

```python
tmp = crime_station.columns.get_level_values(0) + crime_station.columns.get_level_values(1) 

print(tmp) #get_level_values(멀티인덱스 차원(level)) ->>multiindex([()]]인덱스 형태를 합쳐서 하나의 column 단일 인덱스로 사용하고 싶을 때 사용하면됨.
#레벨에 있는 차원을 가져와서 인덱스 형태의 자료를 리턴함.

crime_station.columns = tmp

crime_station
```

```python
real_gu = {'종로구': ['종로', '혜화'], '중구': ['남대문', '중부'],'용산구': ['용산'], '성동구': ['성동'],
           '광진구': ['광진'], '동대문구': ['동대문'],'중랑구': ['중랑'], '성북구': ['성북', '종암'],
           '강북구': ['강북'], '도봉구': ['도봉'],'노원구': ['노원'], '은평구': ['서부', '은평'],
           '서대문구': ['서대문'], '마포구': ['마포'], '양천구': ['양천'], '강서구': ['강서'],
           '구로구': ['구로'], '금천구': ['금천'], '영등포구': ['영등포'], '동작구': ['동작'],'관악구': ['관악'],
           '서초구': ['방배', '서초'], '강남구': ['강남', '수서'], '송파구': ['송파'], '강동구': ['강동']}
print(real_gu.items())
```

```python
crime_station.index
```

```python
tmp_gu = []
for station in crime_station.index :
    for gu_name, real_name in real_gu.items():
        if station in real_name:
            tmp_gu.append(gu_name)

print(tmp_gu)
```

```python
crime_station['구'] = tmp_gu

crime_station
```

```python
crime_gu = crime_station.pivot_table(index='구', aggfunc=np.sum)
crime_gu
```
