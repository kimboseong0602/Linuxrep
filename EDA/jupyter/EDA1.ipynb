데이터합치기

```python
left = pd.DataFrame(
    {
        "Key":  ["K0", "K4", "K2", "K3"], 
        "A" : ["A0", "A1", 'A2', 'A3'],
        "B": ["B0", 'B1', 'B2', 'B3'],
    }
)

right = pd.DataFrame(
    {
        "Key":  ["K0", "K1", "K2", "K3"], 
        "C" : ["C0", "C1", 'C2', 'C3'],
        "D": ["D0", 'D1', 'D2', 'D3'],
    }
)
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on='Key')
```

```python
pd.merge(left, right, how="left", on="Key") #how 일단 레프트는 무조건 넣음. 그리고 on으로 겹치는 키는 병합하는 것임.
#how의 조건이 outer면 두 집합 모두, inner는 교집합
```

```python
def drawGraph():
    plt.figure(figsize=(10, 6))
    plt.plot(t, np.sin(t), label='싸인')
    plt.plot(t, np.cos(t), label='코싸인')
    plt.grid()
    plt.legend() #범례임.
    plt.xlabel('시간')
    plt.ylabel('크기')
    plt.title('싸인코싸인 곡선')
    plt.show()
```

```python
import numpy as np

fp1 = np.polyfit(data_result['인구수'], data_result['총계'], 1) #1차식으로 회귀를 해달라는 것. 숫자 1은 1차함수를 찾아달라.
# 즉 인구수와 총계의 상관관계를 나타내는 1차함수를 찾아줘.

fp1
#출력 : 기울기와 y절편이 나온 것임.
```

```python
f1 = np.poly1d(fp1)
#fp1값을 가지고 실제로 1차함수를 만들어줌.
```

```python
f1(400000)#그래서 y값에 대한 예측값이 나온 것임.
```

```python
fx = np.linspace(100000, 700000, 100)#linspace는 숫자공간을 만들어주는데 100000에서부터 700000까지 100개로 나눠서 이공간을 뽑아줘.
```

```python
fx
```

```python
plt.figure(figsize = (4, 4))
plt.scatter(data_result['인구수'], data_result['총계'], s=50)
plt.plot(fx, f1(fx), ls ='dashed', lw = 3, color = 'g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
data_result['오차'] = data_result['총계'] - f1(data_result['인구수'])
#f1에 넣어서 예측값을 빼 실제값에서 예측값을 빼서 오차를 줌
```

```python
df_descending = data_result.sort_values(by = '오차', ascending = False)
df_ascending = data_result.sort_values(by= '오차', ascending = True)
```

```python
df_descending.head()
```

```python
df_ascending.head()
```

```python
from matplotlib.colors import ListedColormap

color_step = ['#e74c3c', '#2ecc71', '#95a5a6', '#2ecc71', '#3498db', '#3498db'] #막대그래프 색상
my_colormap = ListedColormap(color_step)
```

```python
plt.figure(figsize = (10,10))
plt.scatter(data_result['인구수'], data_result['총계'], c=data_result['오차'], s=50, cmap = my_colormap)
plt.plot(fx, f1(fx), ls='dashed', lw= 3, color = 'gray') #dashed -> 점선, lw -> 선두께

for n in range(5):#plt.text(공간)   공간 안에 text넣는거임. 
    plt.text(df_descending['인구수'][n]*1.02,
            df_descending['총계'][n]*0.98,
            df_descending.index[n], fontsize = 15)
    
    plt.text(df_ascending['인구수'][n]*1.02,
            df_ascending['총계'][n]*0.98,
            df_ascending.index[n], fontsize = 15)

plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.colorbar()
plt.grid()
plt.show()
    
```
